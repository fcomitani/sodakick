```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
import sys, getopt
import csv

pd.set_option('display.max_rows', 500)

import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
%matplotlib inline  
import seaborn as sns
sns.set_style("darkgrid")
```

```python
#fixtures table

def get_fixtures(url):
    res = requests.get(url)
    ## The next two lines get around the issue with comments breaking the parsing.
    comm = re.compile("<!--|-->")
    soup = BeautifulSoup(comm.sub("",res.text),'lxml')
    all_tables = soup.findAll("tbody")
    fix_table = all_tables[0]
    return fix_table


def get_frame_fix(features, table):
    pre_df = dict()
    features_wanted = features
    rows = table.find_all('tr')
    for row in rows:
        if(row.find('th',{"scope":"row"}) != None):
            for f in features_wanted:
                cell = row.find("td",{"data-stat": f})
                
                            
                if f=='match_report':    
                    link=cell.find_all('a')
                    for l in link:
                        text=l['href']
                else:
                    a = cell.text.strip().encode()
                    text=a.decode("utf-8")
                
                if(text == ''):
                    text = np.nan
                #if((f!='player')&(f!='nationality')&(f!='position')&(f!='squad')&(f!='age')&(f!='birth_year')):
                #    text = text.replace(',','')
                
                if f in pre_df:
                    pre_df[f].append(text)
                else:
                    pre_df[f] = [text]
    df = pd.DataFrame.from_dict(pre_df)
    return df
    
def get_fixtures_data(path='https://fbref.com/en/comps/11/schedule/Serie-A-Scores-and-Fixtures'):
    fix_tab=get_fixtures(path)
    df = get_frame_fix(fix,fix_tab)
    df.dropna(axis=0, how='all', inplace=True)
    
    #drop matches that don't exist yet
    df.dropna(axis=0, subset=['score'], inplace=True)

    df['score_a']=df['score'].apply(lambda x: int(x.split('–')[0]))
    df['score_b']=df['score'].apply(lambda x: int(x.split('–')[1]))
    df.drop('score',axis=1, inplace=True)
    
    return df
```

```python
#match report table
def get_fixtures_report(url):
    res = requests.get(url)
    ## The next two lines get around the issue with comments breaking the parsing.
    comm = re.compile("<!--|-->")
    soup = BeautifulSoup(comm.sub("",res.text),'lxml')
    all_tables = soup.findAll("tbody")
    
    #print(len(all_tables)) 7*team+ 3 shots (both+team1+team2, dropping the last two)
    fix_table = all_tables[:-2]
    fix_table.append(soup.find_all("div", {"class": "lineup"})[0])
    fix_table.append(soup.find_all("div", {"class": "lineup"})[1])
    
    return fix_table

def get_frame_play_report(table):
    pre_df = dict()
    pre_df['player']=[]
    rows = table.find_all('tr')
    for row in rows:
        if(row.find('th',{"scope":"row"}) != None):
            cell = row.find("th",{"data-stat": "player"})
            a = cell.text.strip().encode()
            text=a.decode("utf-8")
            pre_df['player'].append(text)     
           
            for cell in row.find_all("td"):
                f = cell['data-stat']
                a = cell.text.strip().encode()
                text=a.decode("utf-8")

                if(text == ''):
                    text = np.nan

                if f in pre_df:
                    pre_df[f].append(text)
                else:
                    pre_df[f] = [text]
                
    df = pd.DataFrame.from_dict(pre_df)
    return df

def get_frame_shots_report(table):
    pre_df = dict()
    pre_df['minute']=[]
    rows = table.find_all('tr')
    for row in rows:
        if(row.find('th',{"scope":"row"}) != None):
            cell = row.find("th",{"data-stat": "minute"})
            a = cell.text.strip().encode()
            text=a.decode("utf-8")
            pre_df['minute'].append(text)     
           
            for cell in row.find_all("td"):
                f = cell['data-stat']
                a = cell.text.strip().encode()
                text=a.decode("utf-8")

                if(text == ''):
                    text = np.nan

                if f in pre_df:
                    pre_df[f].append(text)
                else:
                    pre_df[f] = [text]
                
    df = pd.DataFrame.from_dict(pre_df)
    return df

def get_frame_lineup(table):
    pre_df = dict()
    rows = table.find_all('tr')
    pre_df['player']=[]
    pre_df['team']=[]
    pre_df['bench']=[]
    pre_df['formation']=[]
    for row in rows:
        if(row.find('th') != None):    
            a=row.find('th').text.strip().encode()
            tmp=a.decode("utf-8")
            if tmp=='Bench':
                bench = True
            else:
                bench = False
                team = tmp.split()[0]
                form = tmp.split()[1]
        else:
            a=row.text.strip().encode()
            play=a.decode("utf-8")
            pre_df['player'].append(''.join(i for i in play if not i.isdigit()))
            pre_df['team'].append(team)
            pre_df['bench'].append(bench)
            pre_df['formation'].append(form)

    df = pd.DataFrame.from_dict(pre_df)
    return df

def get_report_data(path='en/matches/2b01c68a/Parma-Juventus-August-24-2019-Serie-A'):
    fix_tab=get_fixtures_report('https://fbref.com/'+path)
        
    team1=pd.concat([get_frame_play_report(fix_tab[x]) for x in range(6)],axis=1)
    team1 = team1.loc[:,~team1.columns.duplicated()]
    team2=pd.concat([get_frame_play_report(fix_tab[x]) for x in range(7,13)],axis=1)
    team2 = team2.loc[:,~team2.columns.duplicated()]
    
    gk1=get_frame_play_report(fix_tab[6])
    gk2=get_frame_play_report(fix_tab[13]) 
    
    shots=get_frame_shots_report(fix_tab[14])
    
    lineup1=get_frame_lineup(fix_tab[15])
    lineup2=get_frame_lineup(fix_tab[16])

    lineup=pd.concat([lineup1,lineup2], axis=0)

    return team1,team2,gk1,gk2,shots,lineup
```

```python
#Functions to get the data in a dataframe using BeautifulSoup

def get_tables(url):
    res = requests.get(url)
    ## The next two lines get around the issue with comments breaking the parsing.
    comm = re.compile("<!--|-->")
    soup = BeautifulSoup(comm.sub("",res.text),'lxml')
    all_tables = soup.findAll("tbody")
    team_table = all_tables[0]
    vs_table = all_tables[1]
    player_table = all_tables[2]
    #all_tables contains 3 objects, changed player_table to 2 rather than 1
    return player_table, team_table, vs_table

def get_frame(features, player_table):
    pre_df_player = dict()
    features_wanted_player = features
    rows_player = player_table.find_all('tr')
    for row in rows_player:
        if(row.find('th',{"scope":"row"}) != None):
    
            for f in features_wanted_player:
                cell = row.find("td",{"data-stat": f})
                a = cell.text.strip().encode()
                text=a.decode("utf-8")
                if(text == ''):
                    text = np.nan
                if((f!='player')&(f!='nationality')&(f!='position')&(f!='squad')&(f!='age')&(f!='birth_year')):
                    try:
                        text = float(text.replace(',',''))
                    except:
                        text = float(text)
                if f in pre_df_player:
                    pre_df_player[f].append(text)
                else:
                    pre_df_player[f] = [text]
    df_player = pd.DataFrame.from_dict(pre_df_player)
    return df_player

def get_frame_team(features, team_table):
    pre_df_squad = dict()
    #Note: features does not contain squad name, it requires special treatment
    features_wanted_squad = features
    rows_squad = team_table.find_all('tr')
    for row in rows_squad:
        if(row.find('th',{"scope":"row"}) != None):
            name = row.find('th',{"data-stat":"squad"}).text.strip().encode().decode("utf-8")
            if 'squad' in pre_df_squad:
                pre_df_squad['squad'].append(name)
            else:
                pre_df_squad['squad'] = [name]
            for f in features_wanted_squad:
                cell = row.find("td",{"data-stat": f})
                a = cell.text.strip().encode()
                text=a.decode("utf-8")
                if(text == ''):
                    text = np.nan
                if((f!='player')&(f!='nationality')&(f!='position')&(f!='squad')&(f!='age')&(f!='birth_year')):
                    try:
                        text = float(text.replace(',',''))
                    except:
                        text = float(text)
                if f in pre_df_squad:
                    pre_df_squad[f].append(text)
                else:
                    pre_df_squad[f] = [text]
    df_squad = pd.DataFrame.from_dict(pre_df_squad)
    return df_squad

def frame_for_category(category,top,end,features):
    url = (top + category + end)
    player_table, team_table, vs_table = get_tables(url)
    df_player = get_frame(features, player_table)
    return df_player

def frame_for_category_team(category,top,end,features):
    url = (top + category + end)
    player_table, team_table, vs_table = get_tables(url)
    df_team = get_frame_team(features, team_table)
    return df_team

def frame_for_category_vs(category,top,end,features):
    url = (top + category + end)
    player_table, team_table, vs_table = get_tables(url)
    df_team = get_frame_team(features, vs_table)
    return df_team

#Function to get the player data for outfield player, includes all categories - standard stats, shooting
#passing, passing types, goal and shot creation, defensive actions, possession, and miscallaneous
def get_outfield_data(top, end):
    df1 = frame_for_category('stats',top,end,stats)
    df2 = frame_for_category('shooting',top,end,shooting2)
    df3 = frame_for_category('passing',top,end,passing2)
    df4 = frame_for_category('passing_types',top,end,passing_types2)
    df5 = frame_for_category('gca',top,end,gca2)
    df6 = frame_for_category('defense',top,end,defense2)
    df7 = frame_for_category('possession',top,end,possession2)
    df8 = frame_for_category('misc',top,end,misc2)
    df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8], axis=1)
    df = df.loc[:,~df.columns.duplicated()]
    return df

#Function to get keeping and advance goalkeeping data
def get_keeper_data(top,end):
    df1 = frame_for_category('keepers',top,end,keepers)
    df2 = frame_for_category('keepersadv',top,end,keepersadv2)
    df = pd.concat([df1, df2], axis=1)
    df = df.loc[:,~df.columns.duplicated()]
    return df


#Function to get team-wise data accross all categories as mentioned above
def get_team_data(top,end):
    df1 = frame_for_category_team('stats',top,end,stats3)
    df2 = frame_for_category_team('keepers',top,end,keepers3)
    df3 = frame_for_category_team('keepersadv',top,end,keepersadv2)
    df4 = frame_for_category_team('shooting',top,end,shooting3)
    df5 = frame_for_category_team('passing',top,end,passing2)
    df6 = frame_for_category_team('passing_types',top,end,passing_types2)
    df7 = frame_for_category_team('gca',top,end,gca2)
    df8 = frame_for_category_team('defense',top,end,defense2)
    df9 = frame_for_category_team('possession',top,end,possession2)
    df10 = frame_for_category_team('misc',top,end,misc2)
    df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10], axis=1)
    df = df.loc[:,~df.columns.duplicated()]
    return df

def get_vs_data(top,end):
    df1 = frame_for_category_vs('stats',top,end,stats3)
    df2 = frame_for_category_vs('keepers',top,end,keepers3)
    df3 = frame_for_category_vs('keepersadv',top,end,keepersadv2)
    df4 = frame_for_category_vs('shooting',top,end,shooting3)
    df5 = frame_for_category_vs('passing',top,end,passing2)
    df6 = frame_for_category_vs('passing_types',top,end,passing_types2)
    df7 = frame_for_category_vs('gca',top,end,gca2)
    df8 = frame_for_category_vs('defense',top,end,defense2)
    df9 = frame_for_category_vs('possession',top,end,possession2)
    df10 = frame_for_category_vs('misc',top,end,misc2)
    df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10], axis=1)
    df = df.loc[:,~df.columns.duplicated()]
    return df
```
